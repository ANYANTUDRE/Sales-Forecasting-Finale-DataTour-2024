```python
train      = pd.read_csv('https://raw.githubusercontent.com/dataafriquehub/donnee_vente/refs/heads/main/train.csv', index_col=0)
submission = pd.read_csv('https://raw.githubusercontent.com/dataafriquehub/donnee_vente/refs/heads/main/submission.csv', index_col=0)

train = train.dropna(subset=['quantite_vendue'])

train      = TabularDataset(train)
submission = TabularDataset(submission)
```

```python
automl = TabularPredictor(label='quantite_vendue', 
                          problem_type='regression', 
                          eval_metric='mean_absolute_percentage_error'
                         )

automl.fit(train_data=train,
           presets='best_quality',
            #keep_only_best = False,
            time_limit= 3600*5,  # 5 hour time limit
            auto_stack=True,
            #num_bag_folds=5,  # Bagging for better generalization
            #num_stack_levels=1,  # Light ensembling to save computation time
            #verbosity=2,
            num_gpus=1,  # Adjust based on availability
          )
```

```python
display(automl.fit_summary())
automl.leaderboard()
```

```python
def post_process(preds):
     return np.array([int(pred) for pred in preds])

def round_predictions(preds):
    return np.array([round(pred) for pred in preds])


preds = automl.predict(submission)
preds_post = post_process(preds)
preds_rounded = round_predictions(preds)

submission_df = pd.DataFrame({
    'id': range(1,len(preds)+1),
    'quantite_vendue': preds
})

submission_df_post = pd.DataFrame({
    'id': range(1,len(preds_post)+1),
    'quantite_vendue': preds_post
})

submission_df_rounded = pd.DataFrame({
    'id': range(1, len(preds_rounded) + 1),
    'quantite_vendue': preds_rounded
})

submission_df.to_csv('submission_autogluon1.csv', index=False)
submission_df_post.to_csv('submission_autogluon1_post.csv', index=False)
submission_df_rounded.to_csv('submission_autogluon1_rounded.csv', index=False)
submission_df_rounded.head()
```
